```python
# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Excel integration
from openpyxl import load_workbook

# Visualization settings
sns.set_style("whitegrid")

# Show charts inside notebook
%matplotlib inline
```

```python
# Creating sample business dataset

business_data = {
    "Month": [
        "January", "February", "March",
        "April", "May", "June"
    ],
    
    "Sales": [
        12000, 15000, 18000,
        17000, 21000, 25000
    ],
    
    "Profit": [
        3000, 4200, 5000,
        4800, 6200, 7500
    ],
    
    "Region": [
        "North", "South", "East",
        "West", "North", "South"
    ]
}

# Converting dictionary into DataFrame
df = pd.DataFrame(business_data)

# Displaying dataset
df
```

```python
# Displaying the first rows of the dataset

df.head()
```

```python
# Displaying structural information about the dataset

df.info()
```

```python
# Generating statistical summary of numerical columns

df.describe()
```

## Business Performance Analysis

```python
# Total sales
total_sales = df["Sales"].sum()

# Total profit
total_profit = df["Profit"].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

## Sales Visualization

```python
# Creating sales visualization

plt.figure(figsize=(8, 5))

sns.barplot(
    x=df["Month"],
    y=df["Sales"]
)

plt.title("Monthly Sales Performance")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

## Sales vs Profit Analysis

```python
# Creating sales and profit comparison chart

plt.figure(figsize=(10, 5))

plt.plot(
    df["Month"],
    df["Sales"],
    marker="o",
    label="Sales"
)

plt.plot(
    df["Month"],
    df["Profit"],
    marker="o",
    label="Profit"
)

plt.title("Sales vs Profit Over Time")

plt.xlabel("Month")
plt.ylabel("Amount")

plt.legend()

plt.show()
```

## Dual Axis Visualization

```python
# Creating dual axis chart

fig, ax1 = plt.subplots(figsize=(10, 5))

# Sales line
ax1.plot(
    df["Month"],
    df["Sales"],
    color="blue",
    marker="o",
    label="Sales"
)

ax1.set_xlabel("Month")
ax1.set_ylabel("Sales", color="blue")

# Creating second axis
ax2 = ax1.twinx()

# Profit line
ax2.plot(
    df["Month"],
    df["Profit"],
    color="orange",
    marker="o",
    label="Profit"
)

ax2.set_ylabel("Profit", color="orange")

# Title
plt.title("Sales vs Profit with Dual Axis")

plt.show()
```

## Regional Sales Analysis

```python
# Grouping sales by region

regional_sales = df.groupby("Region")["Sales"].sum()

print(regional_sales)
```

## Regional Sales Chart

```python
# Creating regional sales chart

plt.figure(figsize=(8, 5))

regional_sales.plot(
    kind="bar",
    color="skyblue"
)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

## Correlation Analysis

```python
# Checking correlation between sales and profit

correlation = df["Sales"].corr(df["Profit"])

print("Correlation between Sales and Profit:", correlation)
```

## Correlation Heatmap

```python
# Creating correlation heatmap

correlation_matrix = df[["Sales", "Profit"]].corr()

plt.figure(figsize=(6, 4))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Exporting Charts

```python
# Exporting chart as image

plt.figure(figsize=(8, 5))

regional_sales.plot(
    kind="bar",
    color="skyblue"
)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.savefig("../images/regional_sales_chart.png")

plt.show()
```

## Exporting Processed Data to Excel

```python
# Exporting dataframe to Excel

df.to_excel("../data/business_dashboard_data.xlsx", index=False)

print("Excel file exported successfully!")
```

## Business Insights and Executive Analysis

```python
# Calculating total sales and profit

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

## Calculating Profit Margin

```python
# Calculating profit margin percentage

profit_margin = (total_profit / total_sales) * 100

print("Profit Margin (%):", round(profit_margin, 2))
```

## Automated Business Insights

```python
# Generating automated business insights

if profit_margin > 25:
    print("The business shows a strong and healthy profit margin.")
else:
    print("The business may need to improve operational efficiency.")
```

## Identifying Top Performing Region

```python
# Finding the region with highest sales

top_region = df.loc[df["Sales"].idxmax()]

print(top_region)
```

## Creating New Business Metrics

```python
# Creating profit margin column

df["Profit Margin (%)"] = (
    df["Profit"] / df["Sales"]
) * 100

df
```

## Profit Margin Visualization

```python
# Visualizing profit margin by month

plt.figure(figsize=(10, 5))

sns.lineplot(
    x=df["Month"],
    y=df["Profit Margin (%)"],
    marker="o"
)

plt.title("Monthly Profit Margin")
plt.xlabel("Month")
plt.ylabel("Profit Margin (%)")

plt.show()
```

## Executive Recommendations

```python
# Generating executive recommendations

if profit_margin > 25:
    recommendation = """
The company demonstrates strong profitability and operational efficiency.
It is recommended to maintain the current growth strategy while exploring
new expansion opportunities in high-performing regions.
"""
else:
    recommendation = """
The company may need to review operational costs and pricing strategies
to improve overall profitability and long-term sustainability.
"""

print(recommendation)
```
